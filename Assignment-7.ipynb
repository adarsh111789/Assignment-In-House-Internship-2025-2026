```python
#1) Practise Pandas Series
#1.1 Create a Pandas Series from Dictionary
import pandas as pd

data = {
    "A": 10,
    "B": 20,
    "C": 30
}

series = pd.Series(data)

print(series)
```

```python
#1.2 Create a Pandas Series from Lists
import pandas as pd

data = [5, 10, 15, 20]

series = pd.Series(data)

print(series)
```

```python
#1.3 Access the Elements of a Series in Pandas
import pandas as pd

data = [100, 200, 300, 400]

series = pd.Series(data)

print(series[0])   # First element
print(series[2])   # Third element
```

```python
#2) DataFrames
#2.1 Make a Pandas DataFrame with a Two-Dimensional Python List
import pandas as pd

data = [
    [1, "Adarsh", 21],
    [2, "Rahul", 22],
    [3, "Aman", 20]
]

df = pd.DataFrame(data, columns=["ID", "Name", "Age"])

print(df)
```

```python
#2.2 Create DataFrame from Python Dict
import pandas as pd

data = {
    "Name": ["Adarsh", "Rahul", "Aman"],
    "Age": [21, 22, 20]
}

df = pd.DataFrame(data)

print(df)
```

```python
#2.3 Create Pandas DataFrame Using List of Lists
import pandas as pd

data = [
    ["Adarsh", 21],
    ["Rahul", 22],
    ["Aman", 20]
]

df = pd.DataFrame(data, columns=["Name", "Age"])

print(df)
```

```python
#2.4 Create a Pandas DataFrame Using List of Tuples
import pandas as pd

data = [
    ("Adarsh", 21),
    ("Rahul", 22),
    ("Aman", 20)
]

df = pd.DataFrame(data, columns=["Name", "Age"])

print(df)
```

```python
#2.5 Create a Pandas DataFrame from List of Dicts
import pandas as pd

data = [
    {"Name": "Adarsh", "Age": 21},
    {"Name": "Rahul", "Age": 22},
    {"Name": "Aman", "Age": 20}
]

df = pd.DataFrame(data)

print(df)
```

```python
#3) Data Iteration
#3.1 Different Ways to Iterate Over Rows in Pandas DataFrame
import pandas as pd

data = {
    "Name": ["Adarsh", "Rahul", "Aman"],
    "Age": [21, 22, 20]
}

df = pd.DataFrame(data)

# Using iterrows()
for index, row in df.iterrows():
    print(index, row["Name"], row["Age"])
```

```python
#3.2 Selecting Rows in Pandas DataFrame Based on Conditions
import pandas as pd

data = {
    "Name": ["Adarsh", "Rahul", "Aman"],
    "Age": [21, 22, 20]
}

df = pd.DataFrame(data)

result = df[df["Age"] > 20]

print(result)
```

```python
#3.3 Select Any Row from a DataFrame Using iloc[]
import pandas as pd

data = {
    "Name": ["Adarsh", "Rahul", "Aman"],
    "Age": [21, 22, 20]
}

df = pd.DataFrame(data)

print(df.iloc[1])
```

```python
#3.4 Limited Rows Selection with Given Column
import pandas as pd

data = {
    "Name": ["Adarsh", "Rahul", "Aman"],
    "Age": [21, 22, 20],
    "City": ["Jaipur", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

print(df.loc[0:1, ["Name", "City"]])
```

```python
#3.5 Drop Rows from DataFrame Based on Certain Condition
import pandas as pd

data = {
    "Name": ["Adarsh", "Rahul", "Aman"],
    "Age": [21, 22, 20]
}

df = pd.DataFrame(data)

df = df[df["Age"] > 20]

print(df)
```

```python
#3.6 Insert Row at Given Position in Pandas DataFrame
import pandas as pd

data = {
    "Name": ["Adarsh", "Rahul"],
    "Age": [21, 22]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    "Name": ["Aman"],
    "Age": [20]
})

df = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print(df)
```

```python
#3.7 Create a List from Rows in Pandas DataFrame
import pandas as pd

data = {
    "Name": ["Adarsh", "Rahul", "Aman"],
    "Age": [21, 22, 20]
}

df = pd.DataFrame(data)

result = df.values.tolist()

print(result)
```
