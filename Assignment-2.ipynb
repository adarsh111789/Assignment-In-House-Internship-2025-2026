```python
# Question 1
# Student Total and Percentage Program

student_name = input("Enter student name: ")
student_class = input("Enter class: ")

maths = float(input("Enter marks for Mathematics: "))
science = float(input("Enter marks for Science: "))
english = float(input("Enter marks for English: "))
computer = float(input("Enter marks for Computer: "))
social_science = float(input("Enter marks for Social Science: "))

total = maths + science + english + computer + social_science
percentage = total / 5

print("\n----- Student Result -----")

print("Student Name:", student_name)
print("Class:", student_class)

print("\nSubject Marks:")
print("Mathematics:", maths)
print("Science:", science)
print("English:", english)
print("Computer:", computer)
print("Social Science:", social_science)

print("\nTotal Marks:", total)
print("Percentage:", percentage)
```

```python
# Question 2
# String Methods Program

string1 = input("Enter first string: ")
string2 = input("Enter second string: ")

# Concatenating strings
combined_string = string1 + " " + string2

print("\nCombined String:", combined_string)

# String Methods
print("\n----- String Methods Output -----")

print("Lower Case:", combined_string.lower())

print("Upper Case:", combined_string.upper())

print("Title Case:", combined_string.title())

print("Swap Case:", combined_string.swapcase())

print("Capitalize:", combined_string.capitalize())

print("Casefold:", combined_string.casefold())

print("Centered String:", combined_string.center(50))

print("Count of 'a':", combined_string.count('a'))

print("Endswith 'a':", combined_string.endswith('a'))

print("Find 'a':", combined_string.find('a'))

print("Is Alphanumeric:", combined_string.isalnum())

print("Is Digit:", combined_string.isdigit())

print("Is Numeric:", combined_string.isnumeric())

print("Is Space:", combined_string.isspace())

print("Replace spaces with '-':", combined_string.replace(" ", "-"))
```

```python
# Question 3
# Practice of Assignment Operators

x = int(input("Enter a number: "))

print("\nInitial Value of x:", x)

# Addition Assignment
x += 5
print("After += 5 :", x)

# Subtraction Assignment
x -= 3
print("After -= 3 :", x)

# Multiplication Assignment
x *= 2
print("After *= 2 :", x)

# Division Assignment
x /= 4
print("After /= 4 :", x)

# Modulus Assignment
x %= 3
print("After %= 3 :", x)
```

```python
# Question 4
# Grade Calculation using Alphabet Grades

student_name = input("Enter student name: ")
student_class = input("Enter class: ")

maths = float(input("Enter marks for Mathematics: "))
science = float(input("Enter marks for Science: "))
english = float(input("Enter marks for English: "))
computer = float(input("Enter marks for Computer: "))
social_science = float(input("Enter marks for Social Science: "))

# Total and Percentage Calculation
total = maths + science + english + computer + social_science
percentage = total / 5

# Alphabet Grade Calculation
if percentage >= 90:
    grade = 'A+'

elif percentage >= 80:
    grade = 'A'

elif percentage >= 70:
    grade = 'B+'

elif percentage >= 60:
    grade = 'B'

elif percentage >= 50:
    grade = 'C'

elif percentage >= 40:
    grade = 'D'

elif percentage >= 33:
    grade = 'E'

else:
    grade = 'Fail'

# Display Result
print("\n----- Student Result -----")

print("Student Name:", student_name)
print("Class:", student_class)

print("\nSubject Marks:")
print("Mathematics:", maths)
print("Science:", science)
print("English:", english)
print("Computer:", computer)
print("Social Science:", social_science)

print("\nTotal Marks:", total)
print("Percentage:", percentage)
print("Grade:", grade)
```
