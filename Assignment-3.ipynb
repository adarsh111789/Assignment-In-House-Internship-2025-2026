```python
# 1) Practice Program for Dictionary, Tuple, and Set
# Dictionary Example
student = {
    "name": "Adarsh",
    "branch": "CSE",
    "year": 2
}
print("Dictionary:")
print(student)

# Accessing values
print(student["name"])
print(student["branch"])

# Adding new value
student["college"] = "SKIT"
print(student)

# Tuple Example
t = (10, 20, 30, 40, 20)
print("\nTuple:")
print(t)

# Accessing elements
print(t[0])
print(t[2])

# Tuple methods
print("Length =", len(t))
print("Count of 20 =", t.count(20))
print("Index of 30 =", t.index(30))

# Set Example
s = {"apple", "banana", "mango"}
print("\nSet:")
print(s)

# Adding element
s.add("orange")
print(s)

# Another set
s2 = {"banana", "grapes"}

# Union
print("Union =", s.union(s2))

# Intersection
print("Intersection =", s.intersection(s2))

# Difference
print("Difference =", s.difference(s2))
```

```python
# 2) Program for Basic Math Operations Using Functions
# Functions for Math Operations
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide(a, b):
    if b == 0:
        return "Division by zero not possible"
    return a / b

# Taking Input from User
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Calling Functions

print("Addition =", add(num1, num2))
print("Subtraction =", subtract(num1, num2))
print("Multiplication =", multiply(num1, num2))
print("Division =", divide(num1, num2))
```

```python
#3) Program to Check Palindrome Number
# Palindrome Number Program
num = int(input("Enter a number: "))
original_num = num
reverse_num = 0

# Reverse the Number
while num > 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num = num // 10
    
# Display Reverse Number
print("Reverse Number =", reverse_num)

# Check Palindrome
if original_num == reverse_num:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```
