```python
# Question 4
nums = (12, 45, 7, 89, 23)

total = 0
highest = nums[0]
lowest = nums[0]

for num in nums:
    total += num

    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Total:", total)
print("Highest:", highest)
print("Lowest:", lowest)
```

```python
# Question 5 

n = (3, 14, 7, 22, 9, 41, 18, 5)

filtered = ()

for num in n:
    if num > 10:
        filtered += (num,)

print("Filtered Tuple:", filtered)
```
