```python
# 1. Convert 1D Array to 2D
import numpy as np

arr = np.array([1,2,3,4,5,6])

arr2d = arr.reshape(2,3)

print(arr2d)
```

```python
# 2. Print Array Attributes
import numpy as np

arr = np.array([[1,2,3],[4,5,6]])

print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
print("Data Type:", arr.dtype)
print("Item Size:", arr.itemsize)
```

```python
# 7. Create a 4×4×3 Array and Extract Value at Second Set, First Row, Last Column
import numpy as np

arr = np.arange(48).reshape(4,4,3)

print(arr)

print("Required Value:")
print(arr[1,0,2])
```

```python
# 8. Create a 4×4 Array and Extract Odd Rows and Even Columns
import numpy as np

arr = np.arange(1,17).reshape(4,4)

print(arr)

print("Odd Rows and Even Columns:")
print(arr[::2,1::2])
```

```python
# 9. Slice First Two Rows and First Two Columns of Second Set from 4×4×3 Array
import numpy as np

arr = np.arange(48).reshape(4,4,3)

print(arr[1,:2,:2])
```

```python
# 10. Replace All Odd Numbers with -1 Using For Loop
import numpy as np

arr = np.array([[23,56,78,93],
                [71,82,13,24]])

for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i,j] % 2 != 0:
            arr[i,j] = -1

print(arr)
```

```python
# 11. Get Indices of Non-Zero Elements
import numpy as np

arr = np.array([1,0,2,0,3,0,4])

indices = np.nonzero(arr)

print(indices)
```
