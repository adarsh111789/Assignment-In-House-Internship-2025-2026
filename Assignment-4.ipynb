```python
#1) Python function to find maximum of three numbers
def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# Input
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

# Function Call
result = maximum(num1, num2, num3)

# Output
print("Maximum number is:", result)
```

```python
#2) Function to return distinct elements from a list
def distinct_list(lst):
    new_list = []

    for i in lst:
        if i not in new_list:
            new_list.append(i)

    return new_list

# Input
numbers = [1, 2, 2, 3, 4, 4, 5]

# Function Call
result = distinct_list(numbers)

# Output
print("Distinct elements:", result)
```

```python
#3) Function to multiply all numbers in a list
def multiply_list(lst):
    result = 1

    for i in lst:
        result = result * i

    return result

# Input
numbers = [2, 3, 4]

# Function Call
answer = multiply_list(numbers)

# Output
print("Multiplication of all numbers:", answer)
```

```python
#4) Function to calculate factorial of a number
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    return fact

# Input
num = int(input("Enter a number: "))

# Function Call
result = factorial(num)

# Output
print("Factorial is:", result)
```

```python
#5) Python program to reverse a string
def reverse_string(text):
    return text[::-1]

# Input
string = input("Enter a string: ")

# Function Call
result = reverse_string(string)

# Output
print("Reversed string is:", result)
```

```python
#6) Function to check whether a number falls within a given range
def check_range(num, start, end):
    if num >= start and num <= end:
        print(num, "is within the range")
    else:
        print(num, "is outside the range")

# Input
number = int(input("Enter number: "))

# Function Call
check_range(number, 1, 10)
```

```python
#7) Function to print even numbers from a given list
def even_numbers(lst):
    print("Even numbers are:")

    for i in lst:
        if i % 2 == 0:
            print(i)

# Input
numbers = [1, 2, 3, 4, 5, 6]

# Function Call
even_numbers(numbers)
```

```python
#8) Function to check whether a number is prime or not
def prime_number(num):

    if num <= 1:
        print(num, "is not a prime number")

    else:
        for i in range(2, num):
            if num % i == 0:
                print(num, "is not a prime number")
                break
        else:
            print(num, "is a prime number")

# Input
number = int(input("Enter a number: "))

# Function Call
prime_number(number)
```

```python
#9) Function to count uppercase and lowercase letters
def count_case_letters(text):

    upper = 0
    lower = 0

    for i in text:
        if i.isupper():
            upper = upper + 1
        elif i.islower():
            lower = lower + 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

# Input
string = input("Enter a string: ")

# Function Call
count_case_letters(string)
```
