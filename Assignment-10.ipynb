```python
# 1. Replace NaN with 0 and interchange rows and columns
import numpy as np

arr = np.array([[6, -8, 73, -110],
                [np.nan, -8, 0, 94]])

print("Original Array:")
print(arr)

# Replace NaN with 0
arr = np.nan_to_num(arr, nan=0)

print("\nAfter Replacing NaN with 0:")
print(arr)

# Interchange rows
rows_swapped = arr[::-1]
print("\nRows Interchanged:")
print(rows_swapped)

# Interchange columns
cols_swapped = arr[:, ::-1]
print("\nColumns Interchanged:")
print(cols_swapped)
```

```python
# 2. Move axes of a 3D array to new positions
import numpy as np

arr3d = np.arange(24).reshape(2, 3, 4)

print("Original Shape:", arr3d.shape)

new_arr = np.moveaxis(arr3d, 0, 2)

print("New Shape:", new_arr.shape)
print(new_arr)
```
