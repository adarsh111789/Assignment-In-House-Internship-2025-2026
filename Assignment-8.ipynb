```python
# 1) Convert a series of date-strings to a timeseries
import pandas as pd

# List of date strings
dates = ['2025-01-01', '2025-02-15', '2025-03-20']

# Convert to datetime series
timeseries = pd.to_datetime(dates)

print(timeseries)
```

```python
#2) DataFrame Merge Operations
#Create Two DataFrames
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aman', 'Riya', 'Karan', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4, 5],
    'Marks': [85, 90, 88, 95]
})

print(df1)
print(df2)
```

```python
#(a) Inner Merge
inner_merge = pd.merge(df1, df2, on='ID', how='inner')

print(inner_merge)
```

```python
#(b) Left Join
left_join = pd.merge(df1, df2, on='ID', how='left')

print(left_join)
```

```python
#(c) Right Join
right_join = pd.merge(df1, df2, on='ID', how='right')

print(right_join)
```

```python
#(d) Index-Based Join using join()
df1_index = df1.set_index('ID')
df2_index = df2.set_index('ID')

join_result = df1_index.join(df2_index)

print(join_result)
```

```python
#(e) Merging with Multiple Keys
df3 = pd.DataFrame({
    'ID': [1, 1, 2, 2],
    'Subject': ['Math', 'Science', 'Math', 'Science'],
    'Marks': [80, 85, 75, 90]
})

df4 = pd.DataFrame({
    'ID': [1, 2, 2],
    'Subject': ['Math', 'Math', 'Science'],
    'Grade': ['A', 'B', 'A']
})

multi_merge = pd.merge(df3, df4, on=['ID', 'Subject'])

print(multi_merge)
```

```python
#3) Concatenation and Merge
#Create Three DataFrames
dfA = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Riya']
})

dfB = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Karan', 'Neha']
})

dfC = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Marks': [78, 88, 91, 85]
})
```

```python
#Vertical Concatenation
concat_df = pd.concat([dfA, dfB])

print(concat_df)
```

```python
#Merge with Third DataFrame
final_merge = pd.merge(concat_df, dfC, on='ID')

print(final_merge)
```
