```python
# 1) CSV File for Address Book
# Import CSV module
import csv

# Open CSV file in write mode
file = open("address_book.csv", "w", newline='')

# Create CSV writer object
writer = csv.writer(file)

# Write column headings
writer.writerow(["Name", "Address", "Mobile", "Email"])

# Taking 3 records from user
for i in range(3):
    print("\nEnter Details of Person", i + 1)

    name = input("Enter Name: ")
    address = input("Enter Address: ")
    mobile = input("Enter Mobile Number: ")
    email = input("Enter Email ID: ")

    # Write data into CSV file
    writer.writerow([name, address, mobile, email])

print("\nData inserted successfully into CSV file.")

# Close file
file.close()
```

```python
#2) DATABASE Practice using SQLite3
#Step 1: Create Database and Tables
import sqlite3

# Connect to database
conn = sqlite3.connect("college.db")

# Create cursor object
cur = conn.cursor()

# Create Student table
cur.execute("""
CREATE TABLE IF NOT EXISTS Student(
    id INTEGER PRIMARY KEY,
    name TEXT,
    branch TEXT
)
""")

# Create Faculty table
cur.execute("""
CREATE TABLE IF NOT EXISTS Faculty(
    fid INTEGER PRIMARY KEY,
    fname TEXT,
    subject TEXT
)
""")

# Create Course table
cur.execute("""
CREATE TABLE IF NOT EXISTS Course(
    cid INTEGER PRIMARY KEY,
    cname TEXT,
    duration TEXT
)
""")

print("Tables created successfully.")

conn.commit()
```

```python
#Step 2: Insert Records
# Insert records into Student table
cur.execute("INSERT INTO Student VALUES(1, 'Adarsh', 'CSE')")
cur.execute("INSERT INTO Student VALUES(2, 'Rahul', 'AI')")
cur.execute("INSERT INTO Student VALUES(3, 'Aman', 'IT')")

# Insert records into Faculty table
cur.execute("INSERT INTO Faculty VALUES(101, 'Sharma', 'Python')")
cur.execute("INSERT INTO Faculty VALUES(102, 'Verma', 'DBMS')")

# Insert records into Course table
cur.execute("INSERT INTO Course VALUES(201, 'Python', '3 Months')")
cur.execute("INSERT INTO Course VALUES(202, 'DBMS', '4 Months')")

print("Records inserted successfully.")

conn.commit()
```

```python
#Step 3: Different SELECT Operations
# Select all students
print("\nStudent Table:")
cur.execute("SELECT * FROM Student")

rows = cur.fetchall()

for row in rows:
    print(row)

# Select specific columns
print("\nStudent Names:")
cur.execute("SELECT name FROM Student")

for row in cur.fetchall():
    print(row)

# Select with condition
print("\nStudents from AI branch:")
cur.execute("SELECT * FROM Student WHERE branch='AI'")

for row in cur.fetchall():
    print(row)
```

```python
#Step 4: Update Data
# Update branch of student
cur.execute("UPDATE Student SET branch='CSE-AI' WHERE id=2")

print("\nData updated successfully.")

conn.commit()
```

```python
#Step 5: Delete Data
# Delete student with id 3
cur.execute("DELETE FROM Student WHERE id=3")

print("\nData deleted successfully.")

conn.commit()
```

```python
#Step 6: Display Final Records
print("\nFinal Student Table:")

cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

# Close connection
conn.close()
```
